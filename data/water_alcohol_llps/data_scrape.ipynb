```python
soup = BeautifulSoup(all_tables, 'html.parser')

# Find the table element
tables = soup.find_all('table')

refs_to_replace = [f'(Ref.{r})' for r in range(200)]

# Iterate through rows in the table
for t, table in enumerate(tables):

	res = []
	for row in table.find_all('tr'):
		# Skip header row if it exists
		if not row.find('th'):
			cells = row.find_all('td')
			# Extract data from each cell and append to the CSV data list
			row_data = []
			for cell in cells:
				el = cell.find('math')
				try:
					exp = float(el.mrow.msup.mrow.mn.text.strip())
					cell_text = float(el.mrow.mn.text.strip())*(10**(-exp))
				except:
					cell_text = cell.get_text(strip=True)
					for ref in refs_to_replace:
						cell_text = cell_text.replace(ref, '')
				row_data.append(cell_text)
			res.append(row_data)

	header_row = table.find('thead').find('tr')
	header_columns =  [th.text.strip().replace('\n','').replace(' ','').replace('.','') for th in header_row.find_all('th')]
	if 'Name' in header_columns:
		df = pd.DataFrame(res, columns=header_columns)
	else:
		df = pd.DataFrame(res)

	if t > 3:
		mol1 = df.iloc[0,0].split(';')[0].replace('(1)', '')
		mol2 = df.iloc[1,0].split(';')[0].replace('(2)', '')
		filename = f'{mol1}_{mol2}.csv'
		header=False
	else:
		filename = f'{t}.csv'
		header=True


	df.to_csv(f'/Users/b324115/Library/CloudStorage/Box-Box/critical_phenomena_saxs/allisons_data/kbi_thermo/water_alcohol_exp/{filename}', index=False, header=header)


```
